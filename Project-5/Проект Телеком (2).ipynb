# Телекоммуникации


<b>Описание проекта: телекоммуникации</b>
<p>Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.</p>
<b>Описание услуг</b>

Оператор предоставляет два основных типа услуг:

- Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
- Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line, «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).
<p><b>Также доступны такие услуги:</b></p>
- Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
- Выделенная линия технической поддержки (TechSupport);
- Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
- Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).
<p>За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.</p>

<b>Описание данных</b>

<p>Данные состоят из файлов, полученных из разных источников:</p>

- contract_new.csv — информация о договоре;
- personal_new.csv — персональные данные клиента;
- internet_new.csv — информация об интернет-услугах;
- phone_new.csv — информация об услугах телефонии.

<p>Во всех файлах столбец customerID содержит код клиента.</p>
<p>Информация о договорах актуальна на 1 февраля 2020.</p>

<p><b>План  работы:</b></p>

1. Загрузка и предварительное ознакомление с данными, содержащимися в ДатаФреймах;
2. Объединение ДатаФреймов и пердобработка данных;
3. Исследовательский анализ данных объединенного ДатаФрейма и удаление лишних признаков, создание новых;
4. Разделение на выборки;
5. Кодирование и масшатабирование признаков (подбор наилучших методово для каждой модели);
6. Построение моделей Логистической регрессии, Случайного Леса и Градиетного бустинга с использованием GridSearch CV;
7. Проверка модели на тестовой выборке;
8. Корректировка признаков в случае не достижения метрики качества и повтор 4,5,6,7 пунктов\ Переход далее;
9. Исследование важности признаков и матрицы ошибок;
10. Итоговый отчет.

<p><b>Основная цель: построение модели, выявляющей клиентов склонных к прекращению использования услуг компании, для их дальнейшего удеражания маркетинговыми средствами.</b></p>

```python
# Подготовим данные для построения корреляционной матрицы Фи
interval_cols = ['monthly_charges', 'total_charges', 'day_with_client']
phik_overview = df_full.drop(['customer_id', 'begin_date', 'end_date','gender','end_date_new'], axis=1)\
.phik_matrix(interval_cols=interval_cols)
# df_full.phik_matrix(interval_cols=interval_cols)
```

Матрица корреляции $\phi_K$ показывает, что у нашего целевого признака самая высокая зависимость с признаками:
- созданным признаком год,
- количество дней с клиентом,
- общие расходы,
- ежемесячные расходы,
- партнер,
- способ оплаты
- далее платные услуги - телевидения, каталог фильмов и другие.

Также есть, связь между признаками:
- ежемесячные расходы и Каталог фильмов, стриминговое ТВ, Мульти линии тел.связи, вид интернет связи и общие расходы. Это связано, со стоимостью услуг, наличие либо отсутствие услуги увеличивает либо уменьшает стоимсоть.
- общие расходы и дней с клиентом, что логично,
- год регистрации и дней с клиентом. Чем дальше год тем больше дней.
- год регистрации и тип оплаты. По диаграмма было видно, что тип оплаты менялся в сторону увеличения количества месячной оплаты и уменьшения годовых и двухлетних контрактов.
- наличие супруги/а и наличие детей. Что также логично.

```python
# Выведем объединенную диаграмму ROC кривой 
def roc_graph(target_train, probabilities):
    fpr, tpr, thresholds = roc_curve(target_train, probabilities)
    plt.figure()
    plt.plot(fpr, tpr, color="darkorange") 
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlim([-0.02, 1.0])
    plt.ylim([0.0, 1.02])
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC-кривая") 
    plt.grid()
    plt.show()
```

```python
def plot_feature_importance(importance,names,model_type):
    
    #Create arrays from feature importance and feature names
    feature_importance = np.array(importance)
    feature_names = np.array(names)
    
    #Create a DataFrame using a Dictionary
    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame(data)
    
    #Sort the DataFrame in order decreasing feature importance
    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)
    
    #Define size of bar plot
    plt.figure(figsize=(10,8))
    #Plot Searborn bar chart
    sns.barplot(x=fi_df['feature_importance'], y=fi_df['feature_names'])
    #Add chart labels
    plt.title(model_type + 'FEATURE IMPORTANCE')
    plt.xlabel('Важность признака')
    plt.ylabel('Названия признаков')
```

```python
# Отобразим feature_importance
plot_feature_importance(model_cat.get_feature_importance(),features_train.columns,'CATBOOST')
```
